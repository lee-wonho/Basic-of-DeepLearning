AND 게이트의 구현 (가중치만 가질 경우)

```python
def AND(x1,x2):
    w1, w2, theta = 0.5, 0.5, 0.7
    tmp = x1*w1 + x2*w2
    if tmp <=theta:
        return 0
    else:
        return 1
```

```python
print(AND(0,0))
print(AND(1,0))
print(AND(1,1))
print(AND(0,1))
```

가중치와 편향을 이용한 AND, NAND, OR 게이트의 구현

```python
def AND(x1,x2):
    x = np.array([x1,x2])
    w = np.array([0.5,0.5])
    b= -0.7
    tmp = np.sum(w*x) +b
    if tmp <= 0:
        return 0
    else:
        return 1
```

```python
def NAND(x1,x2):
    x = np.array([x1,x2])
    w = np.array([-0.5,-0.5])
    b= 0.7
    tmp = np.sum(w*x) +b
    if tmp <= 0:
        return 0
    else:
        return 1
```

```python
def OR(x1,x2):
    x = np.array([x1,x2])
    w = np.array([0.5,0.5])
    b = -0.2
    tmp = np.sum(w*x)+b
    if tmp <= 0 :
        return 0
    else:
        return 1
```

XOR의 구현

```python
def XOR(x1,x2):
    return AND(NAND(x1,x2),OR(x1,x2))
```

```python
print(XOR(0,0))
print(XOR(1,0))
print(XOR(1,1))
print(XOR(0,1))
```
